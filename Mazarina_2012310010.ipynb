```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset Titanic
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

```python
# Melihat 5 baris pertama
print(df.head())

# Statistik deskriptif untuk kolom numerik
print(df.describe())

# Melihat informasi tipe data dan missing value
print(df.info())

# Melihat distribusi kolom kategori
print(df['Survived'].value_counts())
print(df['Pclass'].value_counts())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Distribusi Usia Penumpang Titanic')
plt.xlabel('Usia')
plt.ylabel('Jumlah')
plt.show()
```

Grafik batang ini menunjukkan jumlah penumpang yang selamat dan tidak selamat.

Terlihat bahwa jumlah penumpang yang tidak selamat jauh lebih besar dibanding yang selamat.

📈 Mayoritas penumpang berusia 20–35 tahun.

🔺 Puncak usia penumpang berada di sekitar 25 tahun.

👶 Terdapat cukup banyak anak-anak (usia 0–10 tahun).

👴 Penumpang usia lanjut (>60 tahun) relatif sedikit.

↗️ Distribusi usia miring ke kanan (lebih banyak penumpang muda).

🎯 Menunjukkan bahwa Titanic didominasi oleh penumpang usia produktif.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival berdasarkan Kelas Penumpang')
plt.xlabel('Kelas Penumpang')
plt.ylabel('Jumlah Penumpang')
plt.show()
```

Penumpang kelas 1 memiliki jumlah yang selamat lebih banyak dibanding yang tidak selamat.

Penumpang kelas 2 memiliki jumlah yang hampir seimbang antara yang selamat dan tidak selamat.

Penumpang kelas 3 sebagian besar tidak selamat, dengan jumlah korban paling banyak.

Peluang untuk selamat lebih tinggi pada kelas penumpang yang lebih tinggi.

Kelas sosial berpengaruh terhadap keselamatan, kemungkinan karena akses evakuasi yang lebih baik.
